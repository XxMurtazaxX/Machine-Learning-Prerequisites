# Visualizing correlation of toss to match winnings

```python
toss_and_match_winning_teams = data[data['toss_winner'] == data['winner']]
```

```python
toss_and_match_winning_teams
```

```python
print("Count of the matches won by the teams upon winning the tosses")
toss_and_match_winning_teams_counts = toss_and_match_winning_teams.value_counts('winner')
toss_and_match_winning_teams_counts
```

```python
plt.barh(toss_and_match_winning_teams_counts.index, toss_and_match_winning_teams_counts.values)

plt.xlabel('Winnings <---> Toss Count')
plt.ylabel('Team')
plt.title('Matches Won by Toss-Winning Teams')
plt.show()
```

```python
total_winnings_per_team = data['winner'].value_counts()

plt.barh(total_winnings_per_team.index, total_winnings_per_team.values)
plt.xlabel('Winnings')
plt.ylabel('Team')
plt.title('Visualizing total winnings per team')
plt.show()
```

```python
winning_rate = data['team1'].value_counts() + data['team2'].value_counts()

plt.barh(winning_rate.index, winning_rate.values)
plt.xlabel('Matches')
plt.ylabel('Team')
plt.title('Visualizing total Matches per team')
plt.show()
```

```python
total_matches_per_team = (total_winnings_per_team / total_matches_per_team) * 100

plt.barh(total_matches_per_team.index, total_matches_per_team.values)
plt.xlabel('Winning rate')
plt.ylabel('Team')
plt.title('Winning rate of the team')
plt.show()
```
