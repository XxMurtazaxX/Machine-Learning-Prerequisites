```python
import pandas as pd
```

```python
companies = pd.read_csv('Fortune 500 2017 - Fortune 500.csv')
```

```python
companies.shape
# companies.describe
# companies.info
```

```python
companies.head(5)
```

```python
sectors = companies.groupby('Sector')
```

# How many companies are there in each sector

```python
sectors.size().sort_values(ascending = False)
```

# Leading companies in each sector

```python
sectors.first()  # returns the first company of each group
# sectors.last()   # counter part of df.first()
```

# Get a particular group

```python
sectors.get_group('Apparel')
```

```python
sectors.sum()
```

# Companies with highest revenues

```python
sectors['Revenues'].mean().sort_values(ascending=False)
```

# Companies with highest profits

```python
sectors['Profits'].mean().sort_values(ascending=False)
```
